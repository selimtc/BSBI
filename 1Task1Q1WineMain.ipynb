```python
import pandas as pd
import math

def entropy(series):
    counts = series.value_counts()
    total = len(series)
    return -sum((c/total) * math.log2(c/total) for c in counts)

# parent entropy
H_parent = entropy(df["Decision"])

# split entropy
H_split = df.groupby("Type")["Decision"].apply(
    lambda x: len(x)/len(df) * entropy(x)
).sum()

IG = H_parent - H_split

print("Q1a_Parent Entropy:", H_parent)
print("Child Entropy Type:", H_split)
print("Q1b_Information Gain for Type:", IG)

# Q1a What is the entropy of the parent data in the context of this classification?
# Q1b What is the information gain for splitting on attribute ‘Type’? Show the steps in your calculation.
```

```python
import math
import pandas as pd
from collections import Counter

# Transform the original DataFrame 'df' to match the 'data' list format
# 1. Select and reorder columns
# 2. Map 'Decision' values
# 3. Convert to a list of tuples
df_as_data = df[['Type', 'Price', 'Characteristic', 'Origin', 'Decision']].copy()
df_as_data['Decision'] = df_as_data['Decision'].replace({
    'to buy': 'buy',
    'not to buy': 'not'
})
data_from_df = df_as_data.values.tolist()

# Use the data derived from df instead of the hardcoded list
data = data_from_df

# filter only Type = 'd'
# Note: Assuming 'Type' is the first element (index 0) in the tuples
subset = [row for row in data if row[0] == 'd']

def entropy(counts):
    total = sum(counts.values())
    H = 0
    for c in counts.values():
        p = c / total
        if p > 0:
            H -= p * math.log2(p)
    return H

# parent entropy for Type = d subset
decisions = Counter([row[4] for row in subset]) # Decision is the last element (index 4)
H_parent_subset = entropy(decisions)

attributes = {
    "Price": 1,
    "Character": 2,
    "Origin": 3
} # These indices correspond to the tuple structure (Type, Price, Character, Origin, Decision)

def info_gain(attr_index, current_H_parent, current_subset):
    groups = {}
    for row in current_subset:
        key = row[attr_index]
        groups.setdefault(key, []).append(row[4]) # Decision is the last element (index 4)

    total = len(current_subset)
    H_split = 0
    for key, vals in groups.items():
        counts = Counter(vals)
        H_split += (len(vals) / total) * entropy(counts)

    return current_H_parent - H_split

# compute IG for each remaining attribute
for name, idx in attributes.items():
    print(f"{name} IG = {info_gain(idx, H_parent_subset, subset)}")
```

```python
import math
import pandas as pd
from collections import Counter

# Select needed columns and clean Decision values
df_clean = df[['Type', 'Price', 'Characteristic', 'Origin', 'Decision']].copy()
df_clean['Decision'] = df_clean['Decision'].map({
    'to buy': 'buy',
    'not to buy': 'not'
})

# Convert dataframe to list
data = df_clean.values.tolist()

# Keep only rows where Type == 'd'
subset = [row for row in data if row[0] == 'd']

def entropy(labels):
    counts = Counter(labels)
    total = sum(counts.values())
    return -sum((c / total) * math.log2(c / total) for c in counts.values())

# Parent entropy
parent_entropy = entropy([row[4] for row in subset])

# Attribute indices
attributes = {
    'Price': 1,
    'Characteristic': 2,
    'Origin': 3
}

def information_gain(index):
    total = len(subset)
    weighted_entropy = 0

    for value in set(row[index] for row in subset):
        labels = [row[4] for row in subset if row[index] == value]
        weighted_entropy += (len(labels) / total) * entropy(labels)

    return parent_entropy - weighted_entropy

# Calculate IG
for attr, idx in attributes.items():
    print(f"{attr} IG = {information_gain(idx)}")
```
